```python
import matplotlib.pyplot as plt # For plot
import numpy as np # For arrays (Algebric function, matrices etc.)
import pandas as pd # Data arrangemnet in tabular form (Data managemnet)
%matplotlib inline
```

```python
def Electron_para(phi):
    kB = 1.38e-23 # Value of Boltzmann constant in (SI Unit)
    e = 1.60e-19 # Value of electronic charge.
    phiw = -20 # Value of wall potential
    me = 9.10e-31 # Mass of electron
    ne0 = 1e16 # Initial density of electrons
    eV = 11600 # 1eV equivalent to temperature in Kelvin
    Te = 1*eV
    ne = ne0*np.exp(e*phi/Te/kB) # Formula for electron density
    vce = np.sqrt(2*e*(phi-phiw)/me)/1000 # Cut-off velocity of electron km/s.
    return [ne, vce]
```

```python
EP = Electron_para(-15)
print(EP)
```

```python
phi = [] # Create empty list for potential 
ne = []  # Empty list of electron density
vce = [] # Empty list for the cut-off velocity of electron
for i in range(-15, 1): # Here i is appear for the potential
    EP = Electron_para(i)
    phi.append(i)
    #print(phi)
    ne.append(EP[0])
    #print(ne)
    vce.append(EP[1])
    #print(vce)
```

- Now we plot the electron desnity Vs potential.

```python
#print(phi,ne,vce)
plt.figure(figsize = [7,6])
plt.plot(phi,ne, color = 'b', linestyle = '-')
plt.axis([-20, 0, 0, 1e16])
plt.grid(True)
plt.title('Electron dnesity graph Vs Potential', fontsize = 14)
plt.xlabel("Potential ($\phi$, V)", fontsize = 16, color = "black")
plt.ylabel("Electron density ($n_{e}$, m$^{-3}$)", fontsize = 16, color = "black")
plt.savefig("Electron_density.jpg") # We can use different formats of figure while save it.
```

```python
Data ={}
Data.update({"Potentia":phi, "Electron_density":ne})
#print(Data)
DF= pd.DataFrame(Data)
#print(DF)
DF.to_csv('Electron_density.csv')
```

- Now we plot the electron velocity Vs Potential.

```python
plt.plot(phi, vce, color = 'r', linestyle = '-')
#plt.axis([-20, 0, 0, 5])
plt.grid(True)
plt.title('Cut-off velocity of electrons graph Vs Potential', fontsize = 14)
plt.xlabel("Potential ($\phi$, V)", fontsize = 16, color = "black")
plt.ylabel("Cut-off velocity ($v_{ce}$, km/s)", fontsize = 16, color = "black")
plt.savefig("Electron_velocity.jpg") # We can use different formats of figure while save it.
```

```python
Data ={}
Data.update({"Potentia":phi, "Cut-off_velocity":vce})
#print(Data)
DF= pd.DataFrame(Data)
#print(DF)
DF.to_csv('Electron_velocity.csv')
```
